## Style Sheets

Matplotlib has [style sheets](http://matplotlib.org/gallery.html#style_sheets) you can use to make your plots look a little nicer. They basically create a set of style rules that your plots follow. 

Here is how to use them.

**Before plt.style.use() your plots look like this:**

```python
df1['A'].hist(color = 'grey')
```

Call the style:

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Now our plots look like this:

```python
df1['A'].hist(color = 'grey')
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

## Hexagonal Bin Plot

Useful for Bivariate Data, alternative to scatterplot:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='pink')
```

## Pie plot

we can create a pie plot with DataFrame.plot.pie() or Series.plot.pie(). If our data includes any NaN, they will be automatically filled with 0. A ValueError will be raised if there are any negative values in your data.

```python
series = pd.Series(3 * np.random.rand(4), index=["a", "b", "c", "d"], name="series")

series.plot.pie(
    labels=["A", "B", "C", "D"],
    colors=["r", "g", "b", "c"],
    autopct="%.2f",
    fontsize=10,
    figsize=(6, 6),
    )
   
```

## Bootstrap plot

Bootstrap plots are used to visually assess the uncertainty of a statistic, such as mean, median, midrange, etc.

```python
plt.figure(figsize = (10,8))

from pandas.plotting import bootstrap_plot
data = pd.Series(np.random.rand(1000))
bootstrap_plot(data, size=50, samples=500, color="grey");
```
