# Finance Data Project - EDA

We'll focus on bank stocks and see how they progressed throughout the [financial crisis](https://en.wikipedia.org/wiki/Financial_crisis_of_2007%E2%80%9308) all the way to early 2016.
This dataset has been taken from kaggle.

```python
from pandas_datareader import data, wb    #Pandas datareader allows you to read stock information directly from the internet 
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

**Data**

```python
bank_stocks = pd.read_pickle('all_banks')
```

```python
bank_stocks.head()
```

```python
#max Close price for each bank's stock throughout the time period
bank_stocks['BAC']['Close'].max()
```

```python
bank_stocks.xs(key='Close', axis=1, level = 'Stock Info').max()
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
returns = pd.DataFrame()
```

 We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for 
loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

**pairplot**

```python
#returns[1:]
import seaborn as sns
sns.pairplot(returns[1:])
```

```python
# Worst Drop (4 of them on Inauguration day)
returns.idxmin()
```

```python
# Best Single Day Gain
# citigroup stock split in May 2011, but also JPM day after inauguration.
returns.idxmax()
```

**standard deviation of the returns**

```python
returns.std() # Citigroup riskiest
```

```python
returns.loc['2015-01-01':'2015-12-31'].std() # Very similar risk profiles, but Morgan Stanley or BofA
```

** Creating a distplot using seaborn of the 2015 returns for Morgan Stanley **

```python
sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS Return'],color='green',bins=100)
```

** Creating a distplot using seaborn of the 2008 returns for CitiGroup **

```python
sns.distplot(returns.loc['2008-01-01':'2008-12-31']['C Return'],color='red',bins=100)
```
